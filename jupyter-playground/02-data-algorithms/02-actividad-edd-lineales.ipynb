Usted es trabajador de uber, y le han pasado el siguiente dataset viajes de Uber en 2024. Se le ha proporionado la función `readFile` que lee el archivo 'ncr_ride_bookings.csv' y le entrega una lista con los nombres de las columnas, y una lista con cada una de las filas del dataset, tal como lo puede ver en el código de a continuación.


```python
import csv

def readFile(fileName='data/raw/uber.csv'):
    with open(fileName, mode='r') as file:
        reader = csv.reader(file)
        columnsName = next(reader) # Leer la primera fila como encabezado
        data = [row for row in reader]  # Leer el resto de las filas como datos
        return columnsName, data

columnsName, data = readFile()
print("columnsName:\n", columnsName)
print("First 5 rows of data:")
for i, row in enumerate(data[:5]):
    print(i, row)
```

## Desarrollo de la Actividad
Implemente las siguientes funciones:
1. `cleanDataset`: el cual limpia el conjunto de datos, eliminando todas filas que cumplas alguna de las siguientes condiciones:
    * Estado distinto a "Completed" (es decir, debe dejar solo las filas con estado "Completed").
    * Que tengan "Booking Value", "Ride Distance" o "Driver Ratings" en "null".
    Puede añadir los parámetros que considere necesarios.

La función debe retornar una lista con los datos limpios (puede ser una lista nueva).\
El tiempo de ejecución de su función debe ser $O(n)$, donde $n$ es la cantidad de filas del dataset.

2. Implemente la función `avgColumns`, que retorne una lista con el promedio de las columnas: "Booking Value", "Ride Distance" y "Driver Ratings", de los datos ya limpios.\
*Nota*: debe retornar los datos en la lista en el orden descrito anteriormente.

```python
# Funcion cleanDataset

def cleanDataset(data):

    cleaned_data = []

    for row in data:
        status   = row[columnsName.index('Booking Status')]
        value    = row[columnsName.index('Booking Value')]
        distance = row[columnsName.index('Ride Distance')]
        rating   = row[columnsName.index('Driver Ratings')]

        is_completed = (status == 'Completed')
        has_no_nulls = 'null' not in [value, distance, rating]

        if is_completed and has_no_nulls:
            cleaned_data.append(row)

    return cleaned_data



clean_data = cleanDataset(data)

print(f'Cantidad total de filas: {len(data)}')

print(f'Porcentaje de filas eliminadas: {(len(data) - len(clean_data)) / len(data) * 100:.0f}%')

print(f'Cantidad de filas limpias: {len(clean_data)}')
```

```python
# Funcion avgColumns

def avgColumns(cleaned_data):

    total_rows = len(cleaned_data)

    if total_rows == 0:
        return "No hay datos."

    sum_value = 0.0
    sum_distance = 0.0
    sum_rating = 0.0

    for row in cleaned_data:
        # Sumamos convirtiendo a decimal (float) de forma clara
        sum_value    += float(row[columnsName.index('Booking Value')])
        sum_distance += float(row[columnsName.index('Ride Distance')])
        sum_rating   += float(row[columnsName.index('Driver Ratings')])

    # Calculamos los promedios finales
    avg_value    = sum_value / total_rows
    avg_distance = sum_distance / total_rows
    avg_rating   = sum_rating / total_rows

    return [avg_value, avg_distance, avg_rating]

promedios = avgColumns(clean_data)
print(f'Promedio de Booking Value: {promedios[0]}')
print(f'Promedio de Ride Distance: {promedios[1]}')
print(f'Promedio de Driver Ratings: {promedios[2]}')
```
