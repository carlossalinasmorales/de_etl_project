## API

#### Las **APIs** (Application Programming Interfaces) son conjuntos de reglas y protocolos que permiten que dos aplicaciones o sistemas se comuniquen entre sí.

Las APIs actúan como puentes que facilitan el intercambio de datos y funcionalidades entre diferentes programas.

### ¿Por que son utiles las Apis?
- **Automatización:** Permiten que aplicaciones y servicios trabajen juntos sin intervención humana.
- **Integración:** Facilitan que diferentes sistemas se comuniquen y compartan datos.
- **Escalabilidad:** Permiten que los desarrolladores añadan funcionalidades a aplicaciones sin tener que reinventar la rueda.
- **Eficiencia:** Reducen el tiempo y el costo al proporcionar soluciones reutilizables.

## ¿Qué es HTTP y por qué es importante?

HTTP (Hypertext Transfer Protocol) es el protocolo de comunicación en el que se basa la web.<br>
Es el conjunto de reglas que define cómo los clientes (como tu navegador o un programa de Python) se comunican con los servidores para intercambiar datos. Los métodos HTTP más comunes son:

- GET: Recuperar datos (leer).
- POST: Enviar datos para que sean procesados (crear).
- PUT: Actualizar datos.
- DELETE: Eliminar datos.

Con `requests`, Python permite interactuar con sitios web y APIs usando estos métodos.

# Flujo básico de una API

## Esquema general

Cliente → Request → Servidor → Response (JSON)

---

### 1. Cliente
Aplicación que inicia la solicitud  
Ejemplos:
- Python (requests)
- Navegador
- App móvil

⬇️

### 2. Request (Solicitud HTTP)
El cliente envía una petición al servidor

Incluye:
- URL (endpoint)
- Método (GET, POST, etc.)
- Parámetros (opcional)

Ejemplo:
GET https://api.agify.io/?name=jaime

⬇️

### 3. Servidor
Procesa la solicitud:
- Recibe la request
- Ejecuta lógica (consulta BD, modelo, etc.)
- Prepara la respuesta

⬇️

### 4. Response (JSON)
El servidor responde con datos estructurados

Ejemplo:
```json
{
  "name": "jaime",
  "age": 80,
  "count": 1234
}

### Atributos de Requests

Los navegadores web envían las solicitudes al servidor HTTP y este les responde con el envío de la estructura de la página web:

**status_code:**<br>
> Indica el resultado de la solicitud realizada (como 200 para éxito, 404 para no encontrado, etc.)

**headers:**<br>
>Devuelve los encabezados de la respuesta HTTP en forma de diccionario.<br> 
>Los encabezados contienen información sobre la respuesta, como el tipo de contenido

**text:**<br>
>Devuelve el contenido de la respuesta en formato de texto.<br>
>Si el servidor devuelve datos en HTML, JSON o algún otro formato, este se mostrará como una cadena de texto.

**json():**<br>
>Devuelve el contenido de la respuesta en formato de json. ( no siempre disponible)<br>

# JSON

- JSON (**J**ava**S**cript **O**bject **N**otation) es un formato ligero de intercambio de datos que es fácil de leer y escribir tanto para humanos como para máquinas. 
- Originalmente basado en una notación de objetos de JavaScript, JSON es ahora independiente de cualquier lenguaje y se utiliza ampliamente para transmitir datos entre un servidor y una aplicación web.

### JSON se usa comúnmente en APIs web para intercambiar datos. 

## Pokemones

````python
url = "https://pokeapi.co/api/v2/pokemon?limit=10"
headers = {}
response = requests.get(url, headers=headers)
print(response.text)
````

````python
# # Json mas claro

import requests

headers = {}
url = "https://pokeapi.co/api/v2/pokemon?limit=5"
response = requests.get(url, headers=headers)
datos = json.loads(response.text.encode("utf-8"))
pretty_json = json.dumps(datos, indent=8)
print (pretty_json)
````

````python
df = pd.DataFrame(datos["results"])
df
````

### Isapres

````python
import httpx
import pandas as pd

with httpx.Client(http2=True) as client:
    response = client.get("https://api.boostr.cl/isapres.json")
    data = response.json()

if response.status_code == 200 and isinstance(data.get("data"), list):
    df = pd.DataFrame(data["data"])
else:
    df = pd.DataFrame()
    print(
        "API error:",
        data.get("message", response.text),
        "code:",
        data.get("code", response.status_code),
    )
df
````

### Genero por Nombre

https://genderize.io/

````python
import requests

nombre = "jaime"

url = f"https://api.genderize.io?name={nombre}"
response = requests.get(url)

if response.status_code == 200: # exito
    data = response.json()
    print(f"--- Resultados para '{nombre}' ---")
    print(f"Género: {data['gender']}")
    print(f"Probabilidad: {data['probability'] * 100:.1f}%")
    print(f"Muestras analizadas: {data['count']}")
else:
    print(f"Error: {response.status_code}")
````

### Edad por Nombre

````python
import requests

nombre = "jaime"

url = f"https://api.agify.io?name={nombre}"
response = requests.get(url)

if response.status_code == 200: # exito
    data = response.json()
    print(f"--- Resultados para '{nombre}' ---")
    print(f"Edad: {data['age']}")
    print(f"Muestras analizadas: {data['count']}")
else:
    print(f"Error: {response.status_code}")
````
