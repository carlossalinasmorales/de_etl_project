# 🎬 Ejercicio: Actores que trabajaron en Chile  

Una productora internacional ha contratado a diferentes actores de Hollywood para grabar una serie de películas en distintas ciudades de Chile.  
Cada registro en el dataset **`Horas_Trabajadas.csv`** corresponde a un actor trabajando en una ciudad y región de Chile, con las horas trabajadas durante el día y la noche.  

## 📂 Columnas del dataset
- **Nombre**: Nombre y apellido del actor.  
- **Ciudad**: Ciudad donde trabajó el actor.  
- **Region**: Región administrativa de Chile donde se ubica la ciudad.  
- **Horas Dia**: Número de horas trabajadas en el día.  
- **Horas Noche**: Número de horas trabajadas en la noche.  

La productora quiere **analizar la información para optimizar los tiempos de grabación y la distribución de trabajo**.

# 📝 Actividades

## 1. Cargar y explorar datos

- 1.1 Importa la librería pandas y carga el archivo Horas_Trabajadas.csv en un DataFrame.
- 1.2 Muestra las primeras 10 filas.
- 1.3 Verifica cuántos registros y columnas tiene el dataset.
- 1.4 Obtén un listado de las ciudades únicas.
  
## 2. Selección de columnas

- 2.1 Obtén solo la columna Nombre.
- 2.2 Muestra las columnas Nombre y HorasDia.
- 2.3 Filtra solo los registros de la Región Magallanes.
  
## 3. Filtrado de filas

- 3.1 Filtra todos los registros donde HorasDia > 6.
- 3.2 Filtra los actores que viven en Santiago y tienen HorasNoche > 5.
- 3.3 Obtén todos los registros de Natalie Portman.

## 4. Estadísticas básicas

- 4.1 Calcula el promedio de HorasDia y de HorasNoche.
- 4.2 Encuentra el valor máximo y mínimo de horas trabajadas en el día.
- 4.3 Calcula el total de horas (día + noche) de todo el dataset.

## 5. Operaciones con nuevas columnas

- 5.1 Crea una columna llamada Horas_Totales = HorasDia + HorasNoche.
- 5.2 Filtra los registros con Horas_Totales > 12.
- 5.3 Filtrar registros de Cañete con Horas Totales >= 40 

## 6. Agrupaciones

- 6.1 Agrupa por Ciudad y calcula el promedio de Horas Dia.
- 6.2 Agrupa por Región y calcula la suma de Horas Totales.
- 6.3 Encuentra cuál ciudad tiene mayor promedio de Horas Noche.
