# 🎬 Ejercicio: Actores que trabajaron en Chile  

Una productora internacional ha contratado a diferentes actores de Hollywood para grabar una serie de películas en distintas ciudades de Chile.  
Cada registro en el dataset **`Horas_Trabajadas.csv`** corresponde a un actor trabajando en una ciudad y región de Chile, con las horas trabajadas durante el día y la noche.  

## 📂 Columnas del dataset
- **Nombre**: Nombre y apellido del actor.  
- **Ciudad**: Ciudad donde trabajó el actor.  
- **Region**: Región administrativa de Chile donde se ubica la ciudad.  
- **Horas Dia**: Número de horas trabajadas en el día.  
- **Horas Noche**: Número de horas trabajadas en la noche.  

La productora quiere **analizar la información para optimizar los tiempos de grabación y la distribución de trabajo**.

```python
import pandas as pd

ruta_data = "data/raw/Horas_Trabajadas.csv"

df = pd.read_csv(ruta_data, sep=";")

df.head(10)
```

```python
df.info()
```

```python
ciudades_unicas = df["Ciudad"].unique()
print(ciudades_unicas)
```

```python
columna_nombre = df["Nombre"]
print(columna_nombre)
```

```python
nombre_y_horas = df[["Nombre", "HorasDia"]]
print(nombre_y_horas)
```

```python
filtro_por_magallanes = df[df["Region"] == "Magallanes"]
print(filtro_por_magallanes)
```

```python
filtro_horas_mayores_a_6 = df[df["HorasDia"] > 6]
print(filtro_horas_mayores_a_6)
```

```python
filtro_santiago_o_horas_mayores_a_5 = df[(df["Ciudad"] == "Santiago") | (df["HorasDia"] > 5)]
print(filtro_santiago_o_horas_mayores_a_5)
```

```python
filtro_por_natalie_portman = df[df["Nombre"] == "Natalie Portman"]
print(filtro_por_natalie_portman)
```

```python
promedio_horas_dia = df["HorasDia"].mean()
print(promedio_horas_dia)
promedio_horas_noche = df["HorasNoche"].mean()
print(promedio_horas_noche)
```

```python
max_horas_dia = df["HorasDia"].max()
print(max_horas_dia)
min_horas_noche = df["HorasNoche"].min()
print(min_horas_noche)
```

```python
total_horas = df["HorasDia"].sum() + df["HorasNoche"].sum()
print(total_horas)
```

```python
add_horas_totales = df["HorasDia"] + df["HorasNoche"]
df["HorasTotales"] = add_horas_totales

df_horas_totales = df
print(df_horas_totales)
```

```python
filtro_horas_totales_mayores_a_12 = df_horas_totales[df_horas_totales["HorasTotales"] > 12]
print(filtro_horas_totales_mayores_a_12)
```

```python
filtro_canete_o_horas_totales_mayores_a_40 = df_horas_totales[
    (df_horas_totales["Ciudad"] == "Cañete")
    & (df_horas_totales["HorasTotales"] > 40)
]
print(filtro_canete_o_horas_totales_mayores_a_40)
```

```python
df_por_ciudad = df.groupby("Ciudad").agg({"HorasDia": "mean", "HorasNoche": "mean"})
print(df_por_ciudad)
```

```python
ciudad_con_mayor_promedio_horas_noche = df_por_ciudad["HorasNoche"].idxmax()
print(ciudad_con_mayor_promedio_horas_noche)
```

```python
ciudad_con_menos_horas_totales = df_horas_totales.groupby("Ciudad")["HorasTotales"].sum().idxmin()
print(ciudad_con_menos_horas_totales)
```

```python
top_3 = df_horas_totales.groupby("Nombre")["HorasTotales"].sum().nlargest(3).index
print(top_3)
```
