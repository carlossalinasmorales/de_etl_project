```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('8-foto.jpg')  # Cambia por la ruta de tu imagen

# Paso 2: Convertir la imagen en un array de NumPy
image_array = np.array(image)
print("Dimensiones de la imagen:", image_array.shape)

# Paso 3: Convertir la imagen a escala de grises promediando los valores RGB
# Solo es necesario si la imagen es a color (tiene 3 canales en RGB)
if len(image_array.shape) == 3:  # Verificamos que tenga 3 canales
    grayscale_image = np.mean(image_array, axis=2)
else:
    grayscale_image = image_array  # Si ya es en escala de grises

# Paso 4: Mostrar la imagen original y la imagen en escala de grises
plt.figure(figsize=(10, 5))

# Imagen original
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Imagen original")
plt.axis('off')

# Imagen en escala de grises
plt.subplot(1, 2, 2)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Imagen en escala de grises")
plt.axis('off')

plt.show()
```

### Explicación:

**Paso 1:** Utilizamos PIL.Image.open para cargar la imagen.

**Paso 2:** Convertimos la imagen en un array de NumPy usando np.array(). La imagen cargada es representada en forma de una matriz tridimensional, donde las dimensiones corresponden a la altura, el ancho y los canales de color (RGB).

**Paso 3:** Si la imagen tiene 3 canales (rojo, verde, azul), calculamos la escala de grises tomando el promedio de los valores RGB para cada píxel con np.mean(array, axis=2).

**Paso 4:** Usamos matplotlib para mostrar tanto la imagen original como la imagen convertida en escala de grises.
