# ¿Que es un archivo?

Un **archivo** es un conjunto de datos almacenados en disco. En Python, se pueden trabajar principalmente con:  
- **Archivos de texto** (`.txt`, `.csv`, `.log`...)  
- **Archivos binarios** (`.jpg`, `.exe`, `.mp3`...)  

En esta clase trabajaremos con **archivos de texto**.

## Abrir y cerrar archivos
Para abrir un archivo usamos la función `open()`:

```python
# Sintaxis:
archivo = open("nombre.txt", "modo")
```

### Modos más comunes:

- "r" → lectura (read)
- "w" → escritura (write, sobrescribe)
- "a" → agregar al final (append)

### Siempre es recomendable cerrar el archivo al terminar:

```python
# Sintaxis:
archivo.close()
```


# 1. Leer archivos

## 1.1 Leer todo el contenido

````python
archivo = open("7_ejemplo.txt", "r", encoding="utf-8")
contenido = archivo.read()
print(contenido)
archivo.close()
````

## 1.2 Leer línea por línea

````python
archivo = open("7_ejemplo.txt", "r", encoding="utf-8")
for linea in archivo:
    print(linea.strip())  # .strip() elimina saltos de línea
archivo.close()
````

## 1.3 Leer líneas en lista

````python
archivo = open("7_ejemplo.txt", "r", encoding="utf-8")
lineas = archivo.readlines()
print(lineas)
archivo.close()
````

# 2. Escribir archivos

## 2.1 Sobrescribir el archivo

````python
archivo = open("7_ejemplo_nuevo.txt", "w", encoding="utf-8")
archivo.write("Hola mundo\n")
archivo.write("Python es genial\n")
archivo.close()
````

## 2.2 Agregar información

````python
archivo = open("7_ejemplo_nuevo.txt", "a", encoding="utf-8")
archivo.write("Nueva línea agregada\n")
archivo.close()
````

## 3. Usar with open() (mejor práctica)
Python recomienda usar el contexto with porque cierra el archivo automáticamente:

````python
with open("7_ejemplo.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        print(linea.strip())
````

## Ejemplo

### Guardar nombres en un archivo:

````python
nombres = ["Ana", "Juan", "Pedro", "Lucía"]

with open("7_nombres.txt", "w", encoding="utf-8") as archivo:
    for nombre in nombres:
        archivo.write(nombre + "\n")
````

### Leerlos y mostrarlos:

````python
with open("7_nombres.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        print("Nombre:", linea.strip())
````

## 4. Manejo de archivos CSV

Los **CSV (Comma-Separated Values)** son archivos de texto donde los datos están separados por comas (,) u otros delimitadores.

## 4.1 Leer un CSV con la librería estándar csv:

````python
import csv

with open("7_personas.csv", "r", encoding="utf-8") as archivo:
    lector = csv.reader(archivo)
    for fila in lector:
        print(fila)
````

## 4.2 Escribir en un CSV:

````python
import csv

datos = [
    ["Nombre", "Edad", "Ciudad"],
    ["Pedro", 28, "La Serena"],
    ["María", 35, "Antofagasta"]
]

with open("7_nuevas_personas.csv", "w", encoding="utf-8", newline="-") as archivo:
    escritor = csv.writer(archivo)
    escritor.writerows(datos)
````

## 4.3 Agegar lineas a un CSV

````python
import csv

# Nuevos registros que queremos agregar
nuevos_datos = [
    ["Carlos", 40, "Puerto Montt"],
    ["Sofía", 27, "Temuco"]
]

# Abrimos en modo 'a' (append)
with open("7_nuevas_personas.csv", "a", encoding="utf-8", newline="") as archivo:
    escritor = csv.writer(archivo)
    escritor.writerows(nuevos_datos)

print("Nuevas filas agregadas al archivo CSV.")
````

# 5 Lectura eficiente de archivos grandes

Cuando trabajamos con archivos muy grandes (por ejemplo, cientos de MB o varios GB), **no podemos cargar todo el archivo en memoria de una sola vez**.



## 5.1 Leer línea a línea (streaming):

````python
with open("7_ejemplo.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        # Procesar cada línea sin cargar todo en memoria
        print(linea.strip())
````

## 5.2 Uso de readline() (control más fino)

````python
with open("7_ejemplo.txt", "r", encoding="utf-8") as archivo:
    linea = archivo.readline()
    # hghjkhkhjkhhkhj
    while linea:
        print(linea.strip())
        linea = archivo.readline()
````

## Ejemplo readline()

````python
with open("7_ejemplo.txt", "r", encoding="utf-8") as archivo:
    for _ in range(3):  # Leer solo las primeras 3 líneas
        print(archivo.readline().strip())
````
